```python
normalize_dict = {"b'0'":0, "b'1'":1 }

df['label'] = df['class'].apply(lambda x : normalize_dict[x])
df.pop('class')

df.head(10)
```

```python
label_grouped = df.groupby(['label'])[['label']].count()
label_grouped['share']=label_grouped.label/label_grouped.label.sum()
label_grouped[['share']].plot.bar(figsize=(20,10), color='darkcyan', label=label_grouped['share'])
```

```python
label_grouped
```

```python
# Features correlation coefficent to label
fi = pd.DataFrame({'feature': df.columns,
                   'nan_count': df.isna().sum(),
                  }).sort_values('nan_count', ascending = False)

fi = fi[fi.feature!='label']

fi.head(30).plot.barh(x='feature', y ='nan_count', figsize=(20,10))
```

```python
hnr_features = fi.sort_values('nan_count', ascending=False).head(11)['feature']
```

```python
fi = pd.DataFrame({'feature': df.columns,
                   'correlation': df.corr()['label'].abs(),
                   'true_correlation': df.corr()['label'],
                  }).sort_values('correlation', ascending = False)

fi = fi[fi.feature!='Unnamed: 0']
fi = fi[fi.feature!='label']

fi.head(50).plot.barh(x='feature', y ='true_correlation', figsize=(20,20))

```

```python
fi.sort_values('correlation', ascending = False).head(30)[fi.feature.isin(hnr_features)]
```
