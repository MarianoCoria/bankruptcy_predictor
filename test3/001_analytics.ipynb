```python
import pandas as pd
import pickle

from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('./dataset/dataset.csv')
```

```python
normalize_dict = {"b'0'":0, "b'1'":1 }

df['label'] = df['class'].apply(lambda x : normalize_dict[x])
df.pop('class')

df.head(10)
```

```python
label_grouped = df.groupby(['label'])[['label']].count()
label_grouped['share']=label_grouped.label/label_grouped.label.sum()
label_grouped[['share']].plot.bar(figsize=(20,10), color='darkcyan', label=label_grouped['share'])
```

```python
label_grouped
```

```python
# Features correlation coefficent to label
fi = pd.DataFrame({'feature': df.columns,
                   'nan_count': df.isna().sum(),
                  }).sort_values('nan_count', ascending = False)

fi = fi[fi.feature!='label']

fi.head(30).plot.barh(x='feature', y ='nan_count', figsize=(10,10))
```

```python
hnr_features = fi.sort_values('nan_count', ascending=False).head(11)['feature']
```

```python
fi = pd.DataFrame({'feature': df.columns,
                   'correlation': df.corr()['label'].abs(),
                   'true_correlation': df.corr()['label'],
                  }).sort_values('correlation', ascending = False)

fi = fi[fi.feature!='Unnamed: 0']
fi = fi[fi.feature!='label']

fi.head(30).plot.barh(x='feature', y ='true_correlation', figsize=(10,10))

```

```python
fi.sort_values('correlation', ascending = False).head(30)[fi.feature.isin(hnr_features)]
```

```python
fig, ax = plt.subplots(figsize=(10,10))

p95 = df['Attr54'].quantile(.90)
p5 = df['Attr54'].quantile(.05)
df = df[df.Attr54<p95]
df = df[df.Attr54>p5]

p95 = df['Attr2'].quantile(.90)
p5 = df['Attr2'].quantile(.05)
df = df[df.Attr2<p95]
df = df[df.Attr2>p5]

df[df.label==0].plot.scatter(x='Attr54', y= 'Attr2', ax=ax, s=30,color='darkcyan', alpha=0.5)
df[df.label==1].plot.scatter(x='Attr54', y= 'Attr2', ax=ax, s=30, color='red',  alpha=0.75)
```

```python
with open('./dataset/oversampled_dataset', 'rb') as handle:
    oversampled_dataset = pickle.load(handle)
```

```python
# 30% examples in test data
train = oversampled_dataset['train']
test = oversampled_dataset['test']
train_labels = oversampled_dataset['train_labels']
test_labels = oversampled_dataset['test_labels']
features = oversampled_dataset['features']
```

```python
df_os = pd.DataFrame(data=train, columns=features)
df_os['label']=train_labels
```

```python
fig, ax = plt.subplots(figsize=(10,10))

p95 = df_os['Attr54'].quantile(.90)
p5 = df_os['Attr54'].quantile(.05)
df_os = df_os[df_os.Attr54<p95]
df_os = df_os[df_os.Attr54>p5]

p95 = df_os['Attr2'].quantile(.90)
p5 = df_os['Attr2'].quantile(.05)
df_os = df_os[df_os.Attr2<p95]
df_os = df_os[df_os.Attr2>p5]

df_os[df_os.label==0].plot.scatter(x='Attr54', y= 'Attr2', ax=ax, s=30, color='darkcyan', alpha=0.5)
df_os[df_os.label==1].plot.scatter(x='Attr54', y= 'Attr2', ax=ax, s=30, color='red',  alpha=0.5)
```

```python
fi = pd.DataFrame({'feature': df_os.columns,
                   'correlation': df_os.corr()['label'].abs(),
                   'true_correlation': df_os.corr()['label'],
                  }).sort_values('correlation', ascending = False)

fi = fi[fi.feature!='Unnamed: 0']
fi = fi[fi.feature!='label']

fi.head(50).plot.barh(x='feature', y ='true_correlation', figsize=(20,20))
```
